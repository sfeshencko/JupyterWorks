```python

import matplotlib .pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,2,10000)
y1 = np.sin(6*x)-(-0.05844668)*np.sin(6*x)+(-0.0579822)*np.cos(6*x)
y2 = np.sin(6*x)-(-0.0047674129)*np.sin(6*x)+(-0.07497583504231)*np.cos(6*x)
y3 = np.sin(6*x)-(-0.00731909)*np.sin(6*x)+(-0.00729647)*np.cos(6*x)
plt.grid()
plt.plot(x, y1,c='g')
plt.plot(x, y2,c='r')
plt.plot(x, y3,c='y')
plt.legend(['1','2','3'])
```
