```python
import matplotlib .pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,10,50)
y1 = x
y2 = [i**2 for i in x]
plt.figure(figsize=(9,9))
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('dependency : y1 = x, y2 = x^2')
plt.ylabel('y1', fontsize = 14)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.xlabel('x',fontsize = 14)
plt.ylabel('y2', fontsize = 14)
plt.grid(True)

```

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
plt.plot(x, y, c = 'r')
plt.fill_between(x, y)
```

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
plt.plot(x, y, c = 'r')
plt.fill_between(x, y, where = (y > 0.75) | (y < -0.75))
```

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
plt.plot(x, y, c = 'r')
plt.fill_between(x, y, where = (y > 0), color = 'r', alpha =0.3)
plt.fill_between(x, y, where = (y < 0), color = 'g', alpha = 0.3)
plt.grid()
```

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
m_ev_case = [None, 10, (100,30) ,slice(100,400,15),[0,100,200,300],[10, 50, 100]]
fig, ax = plt.subplot(2,3,figsize=(10,7))
axs = [ax[i,j] for i in range(2) for j in range(3)]
for i, case in enumerate(m_ev_case):
    axs[i].set_title(str(case))
    axs[i].plot(x,y,'o',ls = '-', ms = 7, markevery = case)
```
