```python
import numpy as np
import matplotlib .pyplot as plt
%matplotlib inline
def fac(k):
    if k == 0:
        return 1
    if k == 1:
        return 1
    else:
        return k*fac(k-1)
def cnk(n, k):
    s = fac(n)/(fac(k)*fac(n-k))  # k > n
    return s
def lag(z, n):
    sum1 = sum2 = 0.0
    for k in range(n+1):
        if k % 2 == 0:
            sum1 += cnk(n, k)*((pow(z, k)) / float(fac(k)))
        else:
            sum2 += cnk(n, k)*((pow(z, k)) / float(fac(k)))
    return sum1 - sum2
x = np.linspace(-5,5,100)
y = [lag(i,1) for i in x]
y1=[lag(i,1) for i in x]
y2=[lag(i,2) for i in x]
y3=[lag(i,3) for i in x]
y4=[lag(i,4) for i in x]
y5=[lag(i,5) for i in x]
y6=[lag(i,6) for i in x]
y7=[lag(i,7) for i in x]
plt.figure(figsize = (9,9))
plt.plot(x,y,x,y1,x,y2,x,y3,x,y4,x,y5,x,y6,x,y7)
plt.subplot(x,y1)
plt.grid(True)

```
